````python
import os
import json
import pdfplumber
from config import GROQ_API_KEY, PDF_PASSWORD
from groq import Groq

# 1. LEER EL PDF Y EXTRAER TEXTO
pdf_path = "downloads/CartolaCuentaCorrienteNacionalMensual.pdf"

try:
    with pdfplumber.open(pdf_path, password=PDF_PASSWORD) as pdf:
        leer_pdf_texto = ""
        for page in pdf.pages:
            texto_pagina = page.extract_text()
            if texto_pagina:
                leer_pdf_texto += texto_pagina + "\n"
        
        print(f"‚úÖ PDF le√≠do: {len(leer_pdf_texto)} caracteres")
except Exception as e:
    print(f"‚ùå Error: {e}")
    exit()

# 2. ENVIAR A GROQ CON INSTRUCCIONES DE SISTEMA
client = Groq(api_key=GROQ_API_KEY)

chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "system",
            "content": """Eres un asistente experto en an√°lisis de cartolas bancarias chilenas.
IMPORTANTE: Debes responder √öNICAMENTE con un JSON v√°lido, sin texto adicional ni explicaciones.
El JSON debe tener este formato exacto:
{
  "transacciones": [
    {
      "fecha": "YYYY-MM-DD",
      "descripcion": "texto descriptivo",
      "monto": numero_entero,
      "tipo": "ingreso" o "egreso"
    }
  ]
}

REGLAS:
- En Chile los montos usan punto como separador de miles (ej: 1.234.567 = un mill√≥n)
- No hay decimales en pesos chilenos
- "PAGO", "GIRO", "IMPUESTO" = egreso
- "TRANSFERENCIA DESDE", "TRASPASO DE", "ABONO" = ingreso
- Ignora totales, res√∫menes y headers"""
        },
        {
            "role": "user",
            "content": f"Extrae todas las transacciones de esta cartola:\n\n{leer_pdf_texto[:8000]}"
        }
    ],
    model="llama-3.3-70b-versatile",
    temperature=0.1  # Baja temperatura para respuestas consistentes
)

# 3. PROCESAR RESPUESTA JSON
respuesta = chat_completion.choices[0].message.content

try:
    # Extraer JSON de la respuesta (por si viene con markdown)
    if "```json" in respuesta:
        inicio = respuesta.find("```json") + 7
        fin = respuesta.find("```", inicio)
        respuesta = respuesta[inicio:fin].strip()
    
    # Parsear JSON
    datos = json.loads(respuesta)
    
    print(f"\n‚úÖ Transacciones extra√≠das: {len(datos['transacciones'])}\n")
    
    # Mostrar primeras 3 como ejemplo
    for i, t in enumerate(datos['transacciones'][:3], 1):
        print(f"{i}. {t['fecha']} | {t['descripcion'][:40]} | ${t['monto']:,} | {t['tipo']}")
    
    # Guardar JSON procesado
    with open("outputs/transacciones_procesadas.json", "w", encoding="utf-8") as f:
        json.dump(datos, f, indent=2, ensure_ascii=False)
    
    print(f"\nüíæ Guardado en: outputs/transacciones_procesadas.json")
    
except json.JSONDecodeError as e:
    print(f"‚ùå Error parseando JSON: {e}")
    print(f"Respuesta recibida:\n{respuesta}")
````
